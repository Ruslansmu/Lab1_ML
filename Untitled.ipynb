```python
import numpy as np
import math
import csv
import matplotlib.pyplot as plt
import random
```

```python
#завдання (а) Візуалізувати дані з lab1_train.csv

i, variableX, variableY = np.loadtxt('lab_1_train.csv', unpack = True, delimiter =',', skiprows = 1)
N = len(variableX)
    
#openFile('lab_1_train.csv')
print(variableX)
print("----------------------------------------------------------------")
print(variableY)
plt.plot (variableX, variableY, 'go')
```

```python
def predict_func (Etha, x):
    return Etha[0] + Etha[1]*x
```

```python
def loss_Func(weightZero, weightOne, variableX, variableY, N):
    mean_square=0
    for i in range(N):
        mean_square = mean_square + math.pow(predict_func([weightZero, weightOne], variableX[i]) - variableY[i],2)
    result = mean_square/(2*N)
    return result
```

```python
def theoretical_Etha(x_keys, y_keys):
    summ_XY=0
    summ_X=0
    summ_Y=0
    summ_X_square=0
    for i in range(N):
        summ_XY=summ_XY + x_keys[i]*y_keys[i]
        summ_X=summ_X + x_keys[i]
        summ_Y=summ_Y + y_keys[i]
        summ_X_square=summ_X_square + math.pow(x_keys[i],2)
    
    Etha_1 = (summ_XY - (1/N)*summ_X*summ_Y)/(summ_X_square - (1/N)*math.pow(summ_X,2))
    Etha_0 = (summ_Y/N) - Etha_1*(summ_X/N)
    return Etha_1, Etha_0
```

```python
alpha = 1
zeroEtha = random.uniform (0,10)
firstEtha = random.uniform (0,10)
tolerance = 0.000001
previous_Etha = [0,0]
N = len(variableX)
while (math.fabs(previous_Etha[0]-zeroEtha)>tolerance) or (math.fabs(previous_Etha[1]-firstEtha)>tolerance):
    previous_Etha = [zeroEtha,firstEtha]
    summ_zero_Etha_0 = 0
    summ_zero_Etha_1 = 0
    for i in range (N):
        summ_zero_Etha_0 = summ_zero_Etha_0 + (predict_func([zeroEtha, firstEtha], variableX[i]) - variableY[i])
        summ_zero_Etha_1 = summ_zero_Etha_1 + (predict_func([zeroEtha, firstEtha], variableX[i])- variableY[i])*variableX[i]
        lossFuncTrain = loss_Func(zeroEtha, firstEtha, variableX, variableY, N)
        
    zeroEtha = zeroEtha - (alpha/N)*summ_zero_Etha_0
    firstEtha = firstEtha - (alpha/N)*summ_zero_Etha_1
    print('zeroEtha: ', zeroEtha, 'firstEtha: ', firstEtha, 'mean_square: ', lossFuncTrain)
    print()
```

```python
#завдання (c) втрат на тестових даних з lab1_test.csv

k, testVariableX, testVariableY = np.loadtxt('lab_1_test.csv', unpack = True, delimiter =',', skiprows = 1)
N = len(testVariableX)
print(testVariableX)
print("----------------------------------------------------------------")
print(testVariableY)
print("----------------------------------------------------------------")
lossFuncTest = loss_Func(zeroEtha, firstEtha, testVariableX, testVariableY, N)
print("Функцію втрат на тестових даних",lossFuncTest)
```
